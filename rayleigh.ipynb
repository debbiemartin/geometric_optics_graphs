```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 

wavelengths = np.linspace(300, 1000, 1000)
spectrum = 1./(np.power(wavelengths, 4).astype(np.float32)) * pow(10,10) / 2
plt.plot(wavelengths, spectrum, color='black')

y = np.linspace(0, 0.6, 100)
X,Y = np.meshgrid(wavelengths, y)
extent=(np.min(wavelengths), np.max(wavelengths), np.min(y), np.max(y))

plt.imshow(X, clim=(350,820),  extent=extent, cmap=plt.get_cmap('nipy_spectral'), aspect='auto')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')

plt.fill_between(wavelengths, spectrum, 0.61, color='w')
plt.savefig('WavelengthColors.png', dpi=200)

plt.show()
```
